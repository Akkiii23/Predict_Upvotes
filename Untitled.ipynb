```python
# There is no missing numbers in dataset.
# To make the ordinal encoding on tag
s= df['Tag'].unique()
s
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df['Tag'] = lb.fit_transform(df['Tag'])
```

```python
# Converting floats to integer
df['Reputation'] = df['Reputation'].astype(int)
df['Answers'] = df['Answers'].astype(int)
df['Views'] = df['Views'].astype(int)
df['Upvotes'] = df['Upvotes'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.iloc[:,[0,1,2,3,4,5]].values
y = df.iloc[:,6].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,
                                                random_state=0)
```

```python
# Feature Scaling
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr_pred = lr.predict(x_test)
```

```python
lr_pred
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, lr_pred))
print('MSE:', metrics.mean_squared_error(y_test, lr_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, lr_pred)))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```
